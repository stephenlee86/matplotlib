```python
#!/usr/bin/env python
%matplotlib inline 
# a bar plot with errorbars
import numpy as np
import matplotlib.pyplot as plt

params = {
           'legend.linewidth': 5,
           'xtick.major.pad':10,
           'ytick.major.pad':10,
           'xtick.labelsize':20,
           'ytick.labelsize':20,
           'legend.fontsize':20,
           'lines.markersize':12,
           'axes.labelsize':24,
           'axes.text.fontsize':14
           }

plt.rcParams.update(params)
```

```python
N = 5
menMeans = (20, 35, 30, 15, 10)
menStd = (2, 3, 4, 1, 2)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, menMeans, width, color='0.3', yerr=menStd, label='Men', ecolor='black')

womenMeans = (25, 32, 34, 20, 25)
womenStd = (3, 5, 2, 3, 3)
rects2 = ax.bar(ind + width, womenMeans, width, color='0.6', yerr=womenStd, label='Women', ecolor='black')

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_xticks(ind + width)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend(loc='upper right', frameon=False)
# remove the borders
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.tick_params(axis='x', which='both', bottom='on',top='off',labelbottom='on') 
ax.tick_params(axis='y', which='both', right='off',left='on',labelleft='on') 

plt.tight_layout()
plt.show()
```
